```python
import re
import nltk
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
def processbody(sms):
    #remove brackets ,punctutation, quotions marks
    sms = re.sub(r'[^\wd]', ' ', sms)
    return sms.lower()
def removewords(sms):
    maketext=""
    gettoken = tokenizer.tokenize(sms)
    for k in gettoken:
        if k not in stopWords:
            maketext =maketext + str(k) +" "
    return maketext 
```

```python
makedatasets =[]
for i in datasets:
    if i[0]=="spam":
        makedatasets.append([1,removewords(processbody(i[1]))])
    else:
        makedatasets.append([0,removewords(processbody(i[1]))])
```

```python
words_dict ={}
for data in makedatasets:
    gettoken = tokenizer.tokenize(data[1])
    for k in gettoken:
        try:
            words_dict[k]+=1
        except:
            words_dict[k]=1
DICT_LEN = len(words_dict.keys())
print("DICT LENGTH IS = "+str(DICT_LEN))
ID_TO_DICT  = {}
DICT_TO_ID ={}
DICT_LIST = [k for k in words_dict.keys()]
#print(DICT_LIST)
for r in DICT_LIST:
    DICT_TO_ID[r] = DICT_LIST.index(r)
    ID_TO_DICT[DICT_LIST.index(r)] = k
makevectors = np.zeros(DICT_LEN)
print("DICT TO ID ")
print(DICT_TO_ID)
```

```python
databasesms=[]
for data in makedatasets:
    smsdata =  data [1]
    tokensfromsms = tokenizer.tokenize(smsdata)
    makevectors = np.zeros(DICT_LEN)
    gettokenid=[]
    for k in tokensfromsms:
        try:
            getid=DICT_TO_ID[k]
            gettokenid.append(getid)
        except:
            print("found")   
        makevectors[getid]=1
    databasesms.append([data[0],makevectors])
```
